# scatter plot

```python

sns.set_context("talk")  
sns.set_style("whitegrid")
df['genres'] = df['genres'].astype(str).str.split(',')

df_exploded = df.explode('genres')
df_exploded = df_exploded.reset_index(drop=True)

top_genres = df_exploded['genres'].value_counts().head(10).index.to_list()
df_top = df_exploded[df_exploded['genres'].isin(top_genres)]

fig, axes = plt.subplots(1, 2, figsize=(30, 20), sharey=True)
sns.scatterplot(
    data=df_exploded,
    x='numvotes', y='total_gross',
    hue='genres', alpha=0.5, palette='tab20',
    ax=axes[0], legend=True
)
axes[0].set_title('All Genres',fontsize=18)
axes[0].set_xlabel('numvotes',fontsize=14)
axes[0].set_ylabel('total_gross',fontsize=14)

sns.scatterplot(
    data=df_top,
    x='numvotes', y='total_gross',
    hue='genres', alpha=0.6, palette='bright',
    ax=axes[1]
)
axes[1].set_title('Top 10 Genres Only',fontsize=18)
axes[1].set_xlabel('numvotes',fontsize=14)
axes[1].set_ylabel('total_gross',fontsize=14)

fig.suptitle('How numvotes affect total gross across genres', fontsize=16)
plt.tight_layout()
plt.savefig('votes_vs_gross_by_genre.png', dpi=300, bbox_inches='tight')
plt.show()
```

# Observation
There is a noticeable correlation between a film's popularity (number of votes) and its total box office gross. Most high-grossing films also received a large number of votes, especially in the action, adventure, and thriller genres. Conversely, genres like romance, and news tend to remain in the lower popularity and revenue ranges.

It is also important to note that the dense dots at the illustrate that most of those films with lower numvote are prone to lower totalgross e.g romance and news

# Trends over time and by content

```python
df['genres'].value_counts()
```

```python
df['total_gross']=df['total_gross']/1000000
```

```python
df['genres']=df['genres'].astype(str).str.split(',')
df=df.explode('genres')
top_genres= df['genres'].value_counts().head(10).index.to_list()
df=df[df['genres'].isin(top_genres)]
grouped_df=df.groupby(['start_year','genres'])['total_gross'].mean().reset_index()
plt.figure(figsize=(10,8))
sns.barplot(grouped_df,x='start_year',y='total_gross',hue='genres',palette='bright')
plt.title('Average total gross by year and genres')
plt.ylabel('Average_total gross')
plt.legend(title='genres',bbox_to_anchor=(1.05,1),loc='upper left')
plt.tight_layout()
plt.savefig('genre_vs_year.png',bbox_inches='tight')
plt.show()
```

# Observations
The graph above shows the average box office gross have evolved overtime from 2010-2018 for the top 10 most popular genres.
***
Action based films and adventures have consistently given a lead for the highest average gross making them safe genres to fund across region and very ideal for big budgets
